```python
# Student Performance Prediction
Predict final grade and pass/fail using UCI Student Performance dataset.
```

```python
# Load libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import mean_squared_error, f1_score, roc_auc_score
import numpy as np
import joblib
```

```python
# Load Dataset
df = pd.read_csv(r"C:\Users\saffa\uploads\student-ml-project\data\student.csv.csv", sep=";")
df.head()
```

```python
# Clean data & define targets
# create target columns
df["final_grade"] = df["G3"]
df["pass"] = (df["G3"] >= 10).astype(int)

# drop G1, G2, G3
df = df.drop(["G1", "G2", "G3"], axis=1)

# check the cleaned data
df.head()
```

```python
# Split data into train/test
# Features
X = df.drop(["final_grade", "pass"], axis=1)

# Target
y_reg = df["final_grade"]   # for regression
y_clf = df["pass"]          # for classification

# Train/test split for regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_reg, test_size=0.2, random_state=42)

# Train/test split for classification
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X, y_clf, test_size=0.2, random_state=42)

print("Data ready for ML!")
```

```python
# find all string columns
categorical_cols = X.select_dtypes(include=['object']).columns
print("Categorical columns:", categorical_cols)

# convert text columns into numbers
X = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

# now split data again after encoding
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X, y_reg, test_size=0.2, random_state=42
)

X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X, y_clf, test_size=0.2, random_state=42
)
```

```python
# Train models
# simple random forest for regression
reg_model = RandomForestRegressor(n_estimators=200, random_state=42)
reg_model.fit(X_train_reg, y_train_reg)  # train
print("Regression model done")  

# simple random forest for classification
clf_model = RandomForestClassifier(n_estimators=200, random_state=42)
clf_model.fit(X_train_clf, y_train_clf)
print("Classification model done")

```

```python
# Evaluate models
# regression evaluation
y_pred_reg = reg_model.predict(X_test_reg)
rmse = np.sqrt(mean_squared_error(y_test_reg, y_pred_reg))
print("Regression RMSE:", rmse)

# classification evaluation
y_pred_clf = clf_model.predict(X_test_clf)
y_prob_clf = clf_model.predict_proba(X_test_clf)[:,1]
print("Classification F1 Score:", f1_score(y_test_clf, y_pred_clf))
print("Classification ROC-AUC:", roc_auc_score(y_test_clf, y_prob_clf))
```

```python
# Save models
joblib.dump(reg_model, "../regression_model.pkl")  # save regression
joblib.dump(clf_model, "../classification_model.pkl")  # save classification
print("Models saved ✅")
```

```python
import matplotlib.pyplot as plt

# Regression feature importance
feat_imp = reg_model.feature_importances_
feat_names = X.columns
plt.barh(feat_names, feat_imp)
plt.title("Feature importance (Regression)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

importances = reg_model.feature_importances_
features = reg_model.feature_names_in_

df = pd.DataFrame({'feature': features, 'importance': importances})
df = df.sort_values(by='importance', ascending=False)

plt.figure(figsize=(10,6))
plt.barh(df['feature'], df['importance'])
plt.gca().invert_yaxis()
plt.title("Regression Feature Importances")
plt.show()
```
